# Data pipeline

In this notebook the data is collected and preprocessed before exploratory analysis can be performed.

The data is provided from the online Data Science platform kaggle.

## Loading in relevant packages

```python
import pandas as pd
import numpy as np
import kagglehub
import os
```

## Fetching the data from kaggle.com

```python
# Download latest version
path = kagglehub.dataset_download("meruvakodandasuraj/e-commerce-customer-behavior-and-sales-20202026")

print("Path to dataset files:", path)
```

I find the files in the path

```python
files = os.listdir(path)
```

Then the csv-files are identified

```python
path
```

```python
datasets = {csv[:-4]: pd.read_csv(f'{path}\\{csv}', sep = ",") for csv in files if '.csv' in csv}
```

## Inspection of each dataset

```python
for key in datasets.keys():
    print('The dataset ' + key + ' contains the following columns:')
    for col in datasets[key].columns:
        print(col)
    print('\n')
```

We consider the above datasets to be our complete information regarding our loosely defined objective of modelling churn and additionally other relevant analysis.

```python
for key in datasets.keys():
    print(f"The dataset {key} contains {sum(pd.isna(datasets[key]).sum())} missing values")
```

It turns out that 3 out of 4 datasets contains no missing values. The dataset orders contains 15,749 missing values spread across the following columns

```python
orders = datasets["orders"]

print("Number of missings:")
for col in orders.columns:
    print(f"{col}: {sum(pd.isna(orders[col]))}")
```

The only column in orders containing missing is the column customer_ratings.

```python
print(orders["customer_rating"].describe())
```

```python
print('Proportion of missings: ' + str(round(sum(pd.isna(orders["customer_rating"]))/len(orders["customer_rating"])*100,2)) + '%')
```

From the values and the column name it is likely a review going from between 1.5 to 5. It is yet unclear if it is the order, the customer experience, the product or the firm as a whole that is being reviewed, however it is likely a full customer experienced related to the order. 

Missings likely reflect customers who did not give a review of the order or an attempted review got lost in the data. 

This feature is hard imputate a substitute value as the average is a 4 out of 5 and might bias because of customers who had an excellent or a horrible experience are more likely to leave a review.  

For now the data set customers contains naive features we might consider while doing churn modelling hence we look at it first.
